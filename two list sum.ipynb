## 单向链表求和

有两个单向链表A: `1->3->9->8`和B: `1->2`。求两者的和。【不允许使用其他数据结构】

## 基本思路

应用10进制加法原理: 
```
初始carry为0
v = a[i]+b[j]+carry
carry = v // 10
v = v % 10
```

有一个需要注意的点：

A, B并不是等长度的

````python
class Node:
    """
    Node definition
    v - value
    p - pointer to next node, maybe None
    """
    def __init__(self, v=0, p=None):
        """
        v uint
        p a pointer to next Node
        """
        self.v = v
        self.p = p
    
    def set_p(self, p=None):
        self.p = p
        return self
    
    def set_v(self, v):
        self.v = v
        return self
    
    def get_v(self):
        return self.v
    
    def get_p(self):
        return self.p
    
    def has_p(self):
        return self.p is not None

def _add(A, B, alen, blen):
    """
    alen SHOULD BE bigger or equal to blen
    
    :param A - node list
    :param B - node list
    :param alen - A list length
    :param blen - B list length
    """
    if blen == 0:
        return 0
    if alen > blen:
        carry = add(A.p, B, alen - 1, blen)
        v = A.v + carry
        carry = v // 10
        A.v = v % 10
    else:
        carry = add(A.p, B.p, alen - 1, blen - 1)
        v = A.v + B.v + carry
        carry = v // 10
        A.v = v % 10

    return carry

def two_list_sum(A, B):
    """
    return A+B and the result's length
    """
    def length(a):
        """list length"""
        _len = 0
        while a is not None:
            _len += 1
            a = a.p
        return _len
            
    alen = length(A)
    blen = length(B)
    hlen = 0
    
    if alen < blen:
        h = B
        k = A
        hlen = blen
        carry = add(B, A, blen, alen)
    else:
        h = A
        k = B
        hlen = alen
        carry = add(A, B, alen, blen)
    
    if carry > 0:
        k.v = carry
        k.p = h
        h = k
        hlen += 1

    return h, hlen
````

````python
def create_list(nums):
    """A helper funtion for unittests"""
    header = prev = None
    for num in nums:
        p = Node(num)
        if header == None:
            header = prev = p
        else:
            prev.set_p(p)
            prev = p
    return header

alist = [9,9,8,9]
blist = [2,1]

aheader = create_list(alist)
bheader = create_list(blist)

result, length = two_list_sum(aheader, bheader)

p = result
num_str = ''
for i in range(length):
    num_str += '%d' % p.v
    p = p.p

print(num_str)
````
